```python
paris_df['hasYard'] = paris_df['hasYard'].astype('bool')
paris_df['hasPool'] = paris_df['hasPool'].astype('bool')
paris_df['isNewBuilt'] = paris_df['isNewBuilt'].astype('bool')
paris_df['hasStormProtector'] = paris_df['hasStormProtector'].astype('bool')
paris_df['hasStorageRoom'] = paris_df['hasStorageRoom'].astype('bool')
paris_df = paris_df.drop(columns=['basement','attic','garage'])
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
reg = RandomForestRegressor()
```

```python
X = paris_df
y = X.pop('price')
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
reg.fit(X_train,y_train)
```

```python
reg.score(X_train,y_train)
```

```python
def evaluate(model, X, y):
    predictions = model.predict(X)
    errors = abs(predictions - y)
    mape = 100 * np.mean(errors / y)
    accuracy = 100 - mape
    print('Model Performance')
    print('Average Error: {:0.4f} degrees.'.format(np.mean(errors)))
    print('Accuracy = {:0.2f}%.'.format(accuracy))
    
    return accuracy
```

```python
# Evaluating the model performance

evaluate(reg,X_train,y_train)
```

```python
evaluate(reg,X_test,y_test)
```

```python
feature_importance = reg.feature_importances_
```

```python
for i in range(0,len(feature_importance),1):
    new_array = zip(X.columns,feature_importance)
```

```python
paris_df = pd.DataFrame(new_array)
```

```python
#Sorting feature importance

paris_df.sort_values(by=1,ascending=False)
```
